```python
import random
def main():
    score = float(input("Enter score: "))
    result = determine_grades(score)
    print(f"user score is:{score} is {result}")

    if result == "Excellent":
        print("You won a price!")

    random_score = random.randint(0,100)
    random_result = determine_grades(random_score)
    print(f"random: {random_score} = {random_result}")


def determine_grades(score):
    if score < 0 or score > 100:
        return"Invalid score"
    elif score >= 90:
        return "Excellent"
    elif score >= 50:
        return "Passable"
    else:
        return "Bad"

main()
```
