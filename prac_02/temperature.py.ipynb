```python
MENU = """C - Convert Celsius to Fahrenheit
F - Convert Fahrenheit to Celsius
Q - Quit"""
def main():
    print(MENU)
    choice= input(">>> ").upper()
    valid_choice(choice)
    print("Thank you.")

def valid_choice(choice):
    while choice != "Q":
        if choice == "C":
            celsius = float(input("Celsius: "))
            convert_fahrenheit = covert_celsius_to_fahrenheit(celsius)
            print(f"Result: {convert_fahrenheit:.2f} F")
        elif choice == "F":
            fahrenheit = float(input("Fahrenheit : "))
            convert_celsius = covert_fahrenheit_to_celsius(fahrenheit)
            print(f"Result: {convert_celsius:.2f} C")
        else:
            print("Invalid option")
        print(MENU)
        choice = input(">>> ").upper()

def covert_celsius_to_fahrenheit(celsius):
    return celsius * 9.0 / 5 + 32
def covert_fahrenheit_to_celsius(fahrenheit):
    return 5 / 9 * (fahrenheit - 32)

main()
```
